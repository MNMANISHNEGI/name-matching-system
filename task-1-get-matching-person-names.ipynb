```python
from rapidfuzz import process, fuzz

names = [
    "Geetha", "Gita", "Gitu", "Geeta", "Githa",
    "Gethu", "Githu", "Gitika", "Geetika", "Gitanshi",
    "Geetanshi", "Gitanjali", "Geetanjali", "Gitesh",
    "Geetesh", "Giteshwar", "Geeteshwar", "Gitan",
    "Geetan", "Githan", "Gitanaya", "Geetanaya",
    "Geetali", "Gitali", "Gituja", "Geetuja"
]



def name_score(query, choice, **kwargs):

    query = query.lower()
    choice = choice.lower()

   
    if choice.startswith(query):
        return 100

   
    if query in choice:
        return 90

   
    return fuzz.ratio(query, choice)


query = input("Enter name: ")

matches = process.extract(
    query,
    names,
    scorer=name_score,
    limit=5
)

print("\nBest Match:")
print(matches[0])

print("\nRanked Matches:")
for name, score, _ in matches:
    print(name, "->", score)
```
